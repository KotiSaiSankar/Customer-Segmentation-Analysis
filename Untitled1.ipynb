```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("C:/Users/SEC/Downloads/archive/Mall_Customers.csv")
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
data['Gender'] = data['Gender'].map({'Male': 0, 'Female': 1})
```

```python
features = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
sns.pairplot(features)
plt.show()
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
optimal_clusters = 5
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
data['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', palette='viridis', s=100)
plt.title('Customer Segments')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.show()
```

```python
cluster_summary = data.groupby('Cluster').mean()
print(cluster_summary)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Cluster')
plt.title('Number of Customers in Each Cluster')
plt.xlabel('Cluster')
plt.ylabel('Number of Customers')
plt.show()
```

```python
data.to_csv('Customer_Segmentation_Results.csv', index=False)
```
